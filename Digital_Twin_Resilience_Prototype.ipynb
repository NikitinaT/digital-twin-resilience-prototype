# Digital Twin Resilience Prototype (Visual)

This notebook documents the **current simplified state** of the visual prototype for a university-focused Digital Twin resilience concept. It captures three core views (synthetic baseline, transition into adaptive mode, and tier-based activation) that align with the paper narrative, and provides a minimal roadmap for next steps.

**Repository intent:** lightweight, version-controlled artifacts that can evolve into executable prototypes (event schemas, test vectors, simulation stubs).

## 1) Baseline context (synthetic pre-war university network)

## 2) Crisis state and Digital Twin activation (S1 → adaptive control)

This view represents the **operational consequence** of a regional energy grid disruption (S1). The Digital Twin core is activated to support **crisis assessment** and **adaptive control**, producing stabilized targets for key indicators (energy, connectivity, academic continuity, infrastructure integrity).

![](frontend/public/assets/img/1_2.png)

## 3) Adaptive Tier-Based Activation (priority model)

This view introduces a **tiered activation strategy** for restoring university functions under constrained resources.

- **Tier 1 — Critical Functions:** minimal backbone to keep essential operations safe and coordinated.
- **Tier 2 — Operational Continuity:** core academic and service functions needed to maintain continuity.
- **Tier 3 — Limited Operations:** non-critical facilities and services restored later, conditionally.

![](frontend/public/assets/img/1_3.png)

## Current prototype scope

In the current phase, the prototype is intentionally minimal:

1. **Visual narrative artifacts** (the three screenshots above).
2. **Event schema set** (JSON schemas for heterogeneous inputs + one output decision object).
3. **Test pack** (sample valid/invalid JSON events for a test phase).

No causal modeling or full simulation is implemented at this stage; the focus is on **interfaces, contracts, and decision outputs**.

## Next steps (incremental)

Suggested near-term steps for expanding the prototype without adding unnecessary complexity:

1. **Schema-driven validation**: automated validation of all test vectors against JSON Schemas.
2. **Minimal decision stub**: a deterministic rule-based mapper from inputs → output decision (to validate data flow).
3. **Scenario runner**: load a timeline of events (S0→S1→S3) and emit decisions + logs.
4. **Visualization hook**: render indicator trajectories from outputs (simple matplotlib charts).
